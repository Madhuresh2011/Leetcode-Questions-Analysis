```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python

data = pd.read_csv("Leetcode_Questions.csv")
data
```

```python

data.drop(columns = ['Question_No'], inplace = True)
```

```python
data
```

```python
data.head(10)
```

```python
data.info
```

```python
# Removing the percentage symbol
data['Acceptance'] = data['Acceptance'].apply(lambda x: x[:-1])
```

```python
data
```

```python
# Converting to numeric dtype
data['Acceptance'] = pd.to_numeric(data['Acceptance'], errors = 'coerce')
print(data)
```

```python
data.isnull().sum()
```

```python
data.drop(columns = ['Question_Link', 'Solution'], inplace = True)
```

```python
print(data)
```

```python
most_accepted_ques = data.sort_values(by = 'Acceptance', ascending = False).head(100)
most_accepted_ques
```

```python
data['isPremium'] = data['isPremium'].astype(str).replace({
    'True': 'Premium',
    'False': 'Free'
})
print(data)
```

```python
plt.figure(figsize=(8, 6))
plt.pie(data['Difficulty'].value_counts(), 
        labels=data['Difficulty'].unique(), 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=['#ff9999','#66b3ff','yellow'],
        explode = [0, 0, 0.1])

plt.title('Distribution of Question Difficulty', fontsize=16)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(data['isPremium'].value_counts(), 
        labels=data['isPremium'].unique(), 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['#f8c471','#249ee3'],
        explode = [0.1, 0])
plt.title('Distribution of Premium and Non-Premium Users', fontsize=16)
plt.axis('equal')

plt.show()
```

```python
plt.figure(figsize = (15, 5))
sb.countplot(data = data, x = 'Difficulty', hue = 'isPremium')
plt.title("Premium Questions by Difficulty")

sb.despine()
plt.show()
```

```python
most_accepted_qs = data.pivot_table(index = ['isPremium'], values = ['Acceptance'], aggfunc = 'mean').reset_index()
most_accepted_qs
```

```python
plt.figure(figsize = (10, 5))
sb.barplot(data = most_accepted_qs, x = 'isPremium', y = 'Acceptance')
plt.title("Most Accepted Question Type")

sb.despine()
plt.show()
```

```python
most_accepted_by_diff = data.pivot_table(index = ['Difficulty'], values = ['Acceptance'], aggfunc = 'mean').reset_index()
most_accepted_by_diff
```

```python
plt.figure(figsize = (10, 5))
sb.barplot(data = most_accepted_by_diff, x = 'Difficulty', y = 'Acceptance')
plt.title("Most Accepted Question Type")

sb.despine()
plt.show()
```
